```python
def compare_list_ndarray ():
    
    list_1 = [1, "2xdxdxd,3", '4', 5 ,6]
    list_2 = [1, "2xdxdxd,3", '4', 15 ,26]
    print(list_1 + list_2)
    # broadcasting destegi olmadi
    
    
    list_1 = [1, 2, 3, 4]
    list_2 = [1, 2, 3, 4]
    print(list_1 + list_2) #listeleri ard arda birlestiriyor
    print(list_1 + list_2 + [10])  # +10 yapamiyoruz
    # broadcasting destegi yok 
    
    list_3 = np.asarray([1, 2, 3, 4])
    list_4 = np.asarray([1, 2, 3, 4])
    print(list_3 + list_4)
    print(list_3 + list_4 + 10)  # broadcasting
    
compare_list_ndarray()
```

```python
def display_two_image(im_1, im_2):
    plt.subplot(1, 2, 1)
    plt.imshow(im_1)

    plt.subplot(1, 2, 2)
    plt.imshow(im_2)

    plt.show()
    
    
def rotate(im_1):
    # 90 derece saat yonunde cevirmek
    m, n, k = im_1.shape
    new_image = np.zeros((n, m, k), dtype='uint8')
    for i in range(m):
        for j in range(n):
            new_image[j, i] = im_1[i,j]
    return new_image
```

```python
image_1 = plt.imread('resim.jpg')
```

```python
image_2 = rotate(image_1)

display_two_image(image_1, image_2)
```
