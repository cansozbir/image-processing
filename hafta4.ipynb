```python
def convert_rgb_to_gray(im1):
    m=im1.shape[0]
    n=im1.shape[1]
    im2=np.zeros((m,n))
    for i in range(m):
        for j in range(n):
            im2[i,j]=get_distance(im1[i,j,:])
    return im2
def get_distance(v,w=[1/3,1/3,1/3]):
    a,b,c=v[0],v[1],v[2]
    w1,w2,w3=w[0],w[1],w[2]
    d=((a**2)*w1 +
      (b**2)*w2 +
      (c**2)*w3)**.5
    return d
```

```python
def myf(a,b):
    assert a>=0;" intensity positive", "error intensity not positive"
    if(a<=255-b):
        return a+b
    else:
        return 255
myf(3,123)
```
